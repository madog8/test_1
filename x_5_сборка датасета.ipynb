```python
import pandas as pd
```

```python
df_main = pd.read_excel('db_X_5_cyber.xlsx', index_col=0)
```

```python
df_main.rename(columns={'record_9': 'Результат'}, inplace=True) # переименование
```

```python
df_main = df_main[df_main.index.str.contains('\(')] # удаление объектов без "("
```

```python
df_main = df_main[~df_main.index.str.contains('\(ж\)')] # удаление объектов "Ж"
```

```python
df_main.columns
```

```python
df = df_main[['data_game', 'kef_1', 'kef_2', 'tot', 'favorite', 'tot20', 'tot40', 'abs_51', 'tot51_on_game', 'kef_TM', 'разн_тот51-тот', 'diff_tf_t51','Результат' ]]
```

```python
df.columns
```

```python
df = df.dropna(subset=['tot51_on_game']) # удаление пустого значения
```

```python
df = df.dropna(subset=['diff_tf_t51']) # удаление пустого значения
```

```python
for index, row in df.iterrows():
    if row['diff_tf_t51'] < 0:
        df.at[index, 'Результат'] = 1
    else:
        df.at[index, 'Результат'] = 0
    
    
```

```python
for index, row in df.iterrows():   
    try:
        w = row['tot51_on_game'] - row['tot']
        df.at[index, 'разн_тот51-тот'] = w
        print(w)
    except:
        None
```

```python
df.columns
```

```python
df_grouped = df.groupby('разн_тот51-тот').agg({'kef_TM':'sum', 'Результат':'count'}).reset_index()
```

```python
df_grouped 
```

```python
result_df = pd.DataFrame(columns=['разн_тот51-тот', 'сумма_kef_TM'])
```

```python
result_df
```

```python
for value in df['разн_тот51-тот'].unique():
    sum_kef_TM = df[df['разн_тот51-тот'] == value].query('Результат' == 1)['kef_TM'].sum()
    result_df = result_df.append({'разн_тот51-тот': value, 'сумма_kef_TM': sum_kef_TM}, ignore_index=True)
```
